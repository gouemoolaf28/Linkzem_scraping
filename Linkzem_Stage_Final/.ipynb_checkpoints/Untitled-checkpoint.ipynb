```python
import pandas as pd
import numpy as np
import re
df = pd.read_csv('offres.csv')
pd.set_option('display.max_colwidth', None)
# Remove \t characters and replace with space
df['contact'] = df.contact.str.replace(r'\t', '')
# Remove \n characters and replace with space
df['contact'] = df.contact.str.replace(r'\n', '')
# Remove commas and replace with space
df['contact'] = df.contact.str.replace(r',', ' ')
# Retrieve all email addresses
df['contact'] = df.contact.str.findall('(\S+@\S+)')
# Remove square bracket from a list
df['contact'] = df['contact'].str[0]
# Replace NaN values with empty space
df['contact'] = df['contact'].replace(np.nan, '', regex=True)
df
```

```python

df.to_csv("offers_Linkzem.csv", index=False, header=True,encoding='utf-8')
```
